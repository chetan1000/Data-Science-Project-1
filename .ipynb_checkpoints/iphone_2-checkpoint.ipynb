```python
'''PUNCTUATION REMOVAL------'''

iphone_1["Reviews_no_punc"] = ""


def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, "")
    return text
    
for every_review_statement in iphone["Reviews"]:
    iphone_1["Reviews_no_punc"] = iphone_1["Reviews"].apply(remove_punctuations)
    
iphone_1
```

```python
words_per_review = iphone_1["Reviews_no_punc"].apply(lambda x: len(x.split(" ")))
words_per_review.hist(bins = 50)
plt.xlim(0,1000)
plt.xlabel('Review Length (words)')
plt.ylabel('reviews')
plt.show()
```

```python
print('Average words:', words_per_review.mean())
print('Skewness:', words_per_review.skew())
```

```python
'''----------------------------------------------------------TOKENIZATION--------------------------------------------------'''

import nltk
# nltk.download('punkt')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords


def tokenized():
    for i in range(0,4452):
        tok_dat = word_tokenize(iphone_1["Reviews_no_punc"][i])
        print(tok_dat)
```

```python
tokenized()
```

```python
iphone_1["Tokenized_data"] = iphone_1["Reviews_no_punc"].apply(tokenized())
```

```python
'''------------------------------------------------STOPWORDS REMOVAL--------------------------------------------------------'''


stop = set(nltk.corpus.stopwords.words("english"))

def stopwordsremoval(text):
    #stop_words_removed = text.apply(lambda x: [word for word in x if word not in stop])

    stop_words_removed = [word for word in text if word not in stop]
    
    return stop_words_removed
```

```python
iphone_1["Stopwords_removed"] = iphone_1["Tokenized_data"].apply(stopwordsremoval)
```

```python
iphone_2 = iphone_1[["Ratings", "Reviews_no_punc", "Tokenized_data", "Stopwords_removed"]]
iphone_2
```

```python
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer("english")


# Split the sentences to lists of words.
#iphone_2['Stemmed_column'] = iphone_2["Stopwords_removed"].str.split()

iphone_2['Stemmed_column'] = iphone_2["Stopwords_removed"].apply(lambda x: [stemmer.stem(y) for y in x])

pd.set_option('display.max_colwidth', 2)

iphone_2
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer  = WordNetLemmatizer()
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
```

```python
def lemma_text(text):

    return [lemmatizer.lemmatize(w) for w in text]


iphone_2['Lemma_column'] = iphone_2["Stopwords_removed"].apply(lemma_text)
```
