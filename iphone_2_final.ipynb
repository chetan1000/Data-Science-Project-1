```python
'''----------------------------------------------------------TOKENIZATION--------------------------------------------------'''

import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords


def tokenized_1(sentence):
    
    TOK = nltk.word_tokenize(sentence)
    return TOK
```

```python
for every_review_statement in iphone_1["Reviews_no_punc"]:
    iphone_1["Tokenized_data"] = iphone_1["Reviews_no_punc"].apply(tokenized_1)
```

```python
'''------------------------------------------------STOPWORDS REMOVAL--------------------------------------------------------'''

import nltk
nltk.download('stopwords')
stop = set(nltk.corpus.stopwords.words("english"))

def stopwordsremoval(text):


    stop_words_removed = [word for word in text if word not in stop]
    
    return stop_words_removed
```

```python
iphone_1["Stopwords_removed"] = iphone_1["Tokenized_data"].apply(stopwordsremoval)
```

```python
iphone_2 = iphone_1[["Ratings", "Reviews_no_punc", "Tokenized_data", "Stopwords_removed"]]
iphone_2
```

```python
'''----------------------------------------------------STEMMING-----------------------------------------------------------'''
```

```python
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer("english")


# Split the sentences to lists of words.
#iphone_2['Stemmed_column'] = iphone_2["Stopwords_removed"].str.split()

iphone_2['Stemmed_column'] = iphone_2["Stopwords_removed"].apply(lambda x: [stemmer.stem(y) for y in x])

pd.set_option('display.max_colwidth', 2)

iphone_2
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
lemmatizer  = WordNetLemmatizer()
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
```

```python
def lemma_text(text):
    return [lemmatizer.lemmatize(x) for x in text]

for every_observations in iphone_2['Stopwords_removed']:
    iphone_2['lemmatized_column']=iphone_2['Stopwords_removed'].apply(lemma_text)




def lemma_pos(tokens):
    Lemma=nltk.pos_tag(tokens)
    return Lemma

for every_observation in iphone_2['lemmatized_column']:
    iphone_2['lemmatized_pos_column']=iphone_2['lemmatized_column'].apply(lemma_pos)
```
